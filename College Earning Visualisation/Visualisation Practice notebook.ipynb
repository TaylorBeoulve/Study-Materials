```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
```

```python
raw_count = (recent_grads.shape)[0]
recent_grads=recent_grads.dropna()
cleaned_count = (recent_grads.shape)[0]
print(raw_count, cleaned_count)
```

```python
ax1 = recent_grads.plot(x='Sample_size', y='Median', 
                        kind='scatter')
ax1.set_title('Sample Size vs Median')
```

```python
ax2 = recent_grads.plot(x= 'Sample_size', y= 'Unemployment_rate',
                       kind= 'scatter')
ax2.set_title('Sample Size vs Unemployment')
```

```python
ax3 = recent_grads.plot(x = 'Full_time', y= 'Median',
                       kind= 'scatter')
ax3.set_title('Fulltime vs Median')
```

```python
ax4 = recent_grads.plot(x= 'ShareWomen', y= 'Unemployment_rate',
                        kind='scatter')
ax4.set_title('% of Women vs Unemployment')
```

```python
ax5 = recent_grads.plot(x= 'Men', y= 'Median', kind = 'scatter',
                       title='Men vs Median')
```

```python
ax6 = recent_grads.plot(x='Women',y='Median', kind= 'scatter',
                       title='Women vs Median')
```
